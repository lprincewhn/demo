```python
from langchain_openai import OpenAIEmbeddings
from langchain_ollama import OllamaEmbeddings
import os
embeddings_model = OllamaEmbeddings(
    model="bge-m3:567m",
    base_url="http://localhost:11434"
)
```

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader = WebBaseLoader("https://python.langchain.com/docs/how_to/MultiQueryRetriever/")
data = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_splitter.split_documents(data)
vetordb = Chroma.from_documents(documents=splits, embedding=embeddings_model)
```

```python
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain_openai import ChatOpenAI

base_url="https://ark.cn-beijing.volces.com/api/v3"

model="doubao-1.5-pro-32k-250115"
llm = ChatOpenAI(model=model, temperature=0, api_key=api_key, base_url=base_url)

question = "如何让用户查询更准确"
retriever_from_llm = MultiQueryRetriever.from_llm(
    retriever=vetordb.as_retriever(), llm=llm
)
```
