```python
from langchain_openai import OpenAIEmbeddings
from langchain_ollama import OllamaEmbeddings
import os
embeddings_model = OllamaEmbeddings(
    model="bge-m3:567m",
    base_url="http://localhost:11434"
)
```

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader = WebBaseLoader("https://python.langchain.com/docs/how_to/MultiQueryRetriever/")
data = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_splitter.split_documents(data)
vetordb = Chroma.from_documents(documents=splits, embedding=embeddings_model)
```

```python
retriever = vetordb.as_retriever()
docs = retriever.invoke("如何安装langchain")
print(docs)
```
